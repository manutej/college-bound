2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python

#n = len(arr)
#sigma(...) = sum(...)
#xi = element of list
#mu = mean
# sqrt((1/len(arr))*(sum(xi-mean(arr))**2)) --> Translation of formula above to pseudocode, Do not Run

# Code:
def get_stdev(arr):
    n = len(arr)
    mean_value = get_mean(arr)
    sum_arr = []
    for x in arr:
        sum_arr.append((x - mean_value)**2)
    sum_arr = sum(sum_arr)
    
    return (sum_arr*(1/n))**0.5

get_stdev(A)


# TESTED to be correct for multiple inputs of A

# We can verify our answer manually using the following standard deviation calculator link: 
# SOURCE: https://www.mathsisfun.com/data/standard-deviation-calculator.html
```

```python
def plot_corr_heatmap(df,title="Title"):
    # Correlation 
    corr = df.corr()
    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(12, 12))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    sns.set(font_scale=1.5)
    ax.set_title(title)
    # Draw the heatmap with the mask and correct aspect ratio
    return sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, vmin = -1, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})

```

### REFERENCES: 
#### 1) [Plotting a diagonal correlation matrix](https://seaborn.pydata.org/examples/many_pairwise_correlations.html)
#### 2) [seaborn.heatmap documentation](https://seaborn.pydata.org/generated/seaborn.heatmap.html)
#### 3) [How do I add a title and axis labels to Seaborn Heatmap?](https://stackoverflow.com/questions/32723798/how-do-i-add-a-title-and-axis-labels-to-seaborn-heatmap)

### 3. Plot and interpret boxplots. 
    - Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.
    - Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).
    - Each boxplot should:
        - Only include variables of a similar scale
        - Have clear labels for each variable
        - Have appropriate titles and labels

```python
pd.DataFrame(({'Lower Quartile':df_colleges['sat_lower_quartile'],'Upper Quartile':df_colleges['sat_upper_quartile']})).boxplot(
    color = 'steelblue',figsize = (15,10),notch=True, patch_artist=True);
plt.xlabel('Lower and Upper Quartiles', fontsize=20);
plt.ylabel('SAT Combined Scores',fontsize=20);
plt.title("Lower and Upper Quartiles Spread",fontdict = {'fontsize' : 28})
plt.savefig('../images/SAT_quartile_boxplot.png',bbox_inches='tight')
#df_major.info()


```

For this reason it is much easier with a box plot certain key statistical features like median and interquartile range. Since our data set in this graph is already a categorical representation of lower and upper quartile range, we can think of this boxplot as the spread of each range across a range of colleges.
